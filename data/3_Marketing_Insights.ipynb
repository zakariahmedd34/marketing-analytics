#Task 2 - Analytics

- read data_preprocessed.csv
- calculate meaningful patterns in the data
- convert those patterns into plain-English insights
- save each insight as a separate .txt file

```python
import pandas as pd

df_analytics = pd.read_csv("data_preprocessed.csv")
df_analytics.head()
```

```python
print("Shape:", df_analytics.shape)
print("\nColumns:")
print(df_analytics.columns.tolist())
```

##Insight 1 - Income level vs total spending

**Which income level has the highest average total spending?**

We group the data by Income_Level and calculate the mean of Total_Spending.

```python
spending_by_income = (
    df_analytics.groupby("Income_Level")["Total_Spending"]
    .mean()
    .sort_values(ascending=False)
)

spending_by_income
```

```python
top_income_group = spending_by_income.idxmax()
top_income_value = spending_by_income.max()

lowest_income_group = spending_by_income.idxmin()
lowest_income_value = spending_by_income.min()

print("Top income group:", top_income_group)
print("Highest average spending:", round(top_income_value, 2))

print("Lowest income group:", lowest_income_group)
print("Lowest average spending:", round(lowest_income_value, 2))
```

```python
insight1 = (
    f"Insight 1:\n"
    f"Customers in the {top_income_group} income group have the highest average total spending "
    f"({top_income_value:.2f}), while customers in the {lowest_income_group} income group have the lowest "
    f"average total spending ({lowest_income_value:.2f}). "
    f"This suggests that income level is strongly associated with customer spending behavior."
)

print(insight1)
```

##Insight 2 - Age group vs total purchases

**Which age group has the highest average total number of purchases?**

We group the data by Age_Group and calculate the mean of Total_Purchases.

```python
purchases_by_age = (
    df_analytics.groupby("Age_Group")["Total_Purchases"]
    .mean()
    .sort_values(ascending=False)
)

purchases_by_age
```

```python
top_age_group = purchases_by_age.idxmax()
top_age_value = purchases_by_age.max()

lowest_age_group = purchases_by_age.idxmin()
lowest_age_value = purchases_by_age.min()

print("Top age group:", top_age_group)
print("Highest average total purchases:", round(top_age_value, 2))

print("Lowest age group:", lowest_age_group)
print("Lowest average total purchases:", round(lowest_age_value, 2))
```

```python
insight2 = (
    f"Insight 2:\n"
    f"The {top_age_group} age group has the highest average total number of purchases "
    f"({top_age_value:.2f}), while the {lowest_age_group} age group has the lowest "
    f"average total number of purchases ({lowest_age_value:.2f}). "
    f"This suggests that customer age is related to purchase activity."
)

print(insight2)
```

##Insight 3 - Spending level vs response and engagement

**Which spending level has the highest campaign response and engagement?**

We group the data by Spending_Level and calculate:
- average Response
- average Engagement

```python
response_engagement_by_spending = (
    df_analytics.groupby("Spending_Level")[["Response", "Engagement"]]
    .mean()
    .sort_values(by="Response", ascending=False)
)

response_engagement_by_spending
```

```python
top_response_group = response_engagement_by_spending["Response"].idxmax()
top_response_value = response_engagement_by_spending["Response"].max()

top_engagement_group = response_engagement_by_spending["Engagement"].idxmax()
top_engagement_value = response_engagement_by_spending["Engagement"].max()

print("Highest response group:", top_response_group)
print("Highest average response:", round(top_response_value, 4))

print("Highest engagement group:", top_engagement_group)
print("Highest average engagement:", round(top_engagement_value, 4))
```

```python
insight3 = (
    f"Insight 3:\n"
    f"Customers in the {top_response_group} spending group have the highest average response rate "
    f"({top_response_value:.4f}), while customers in the {top_engagement_group} spending group show the highest "
    f"average engagement ({top_engagement_value:.4f}). "
    f"This indicates that customers with stronger spending behavior are also more responsive and engaged in marketing activity."
)

print(insight3)
```

saving insights

```python
with open("insight1.txt", "w") as f:
    f.write(insight1)

with open("insight2.txt", "w") as f:
    f.write(insight2)

with open("insight3.txt", "w") as f:
    f.write(insight3)

print("Saved all insights.")
```
